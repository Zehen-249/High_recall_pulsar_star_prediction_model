```python
from seaborn import heatmap

corr_matrix=dataset.corr()
plt.figure(figsize=(15, 8))
heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

### Every feature has some significant correlation with target_class

```python
import seaborn as sns
for column in dataset.select_dtypes(include=['number']).columns:
    plt.figure(figsize=(10, 6))
    sns.kdeplot(dataset[column], fill=True)
    plt.title(f'Density Plot of {column}')
    plt.show()
```
